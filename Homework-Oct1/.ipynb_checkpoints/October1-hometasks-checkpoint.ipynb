## *Задание №1* (нахождение дискриминанта/решение квадратного уравнения):

```python
from math import sqrt #импортируем функцию квадратного корня из библиотеки math


def extra_num(c, d):
    return c + (d * (-1)) #в случае, если после знака равно есть значение, отличное от нуля, с помощью этой функции мы "переносим" это значение в левую часть


def quadeq(a, b, c):
    if a == 0 and b == 0:
        return "Уравнение не имеет смысла." #с учётом, что к a и b привязан x, то уравнение, где a и b оба равны нулю не имеет смысла
    else:
        D = (b**2) - (4 * a * c) #вычисляем дискриминант
        if D == 0:
            return f"Корень уравнения:{((b * -1) + sqrt(D)) / 2 * a}." #возвращает 1 корень уравнения. конечно, там не 1 корень, а 2 одинаковых, но так ли это в данном случае важно?
        elif D > 0:
            return f"Корни уравнения: {((b * -1) + sqrt(D)) / 2 * a}, {((b * -1) - sqrt(D)) / 2 * a}." #возвращает корни уравнения
        elif D < 0:
            return f"Нет корней." #отображается в случае отсутствия корней


a = float(input("Введите значение постоянной переменной a: "))
b = float(input("Введите значение постоянной переменной b: "))
c = float(input("Введите значение постоянной переменной c: "))
d = float(input("Введите чему равно уравнение: "))
c = extra_num(c, d)
print(quadeq(a, b, c))
```
